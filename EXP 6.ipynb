```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Read the CSV file into a DataFrame
df = pd.read_csv('height-weight.csv')

# Extract the height and weight columns
x = df['height']
y = df['weight']

# Calculate the linear regression coefficients
intercept, slope = np.polyfit(x, y, 1)

# Define a function to predict weight
def pred_y(x):
    return x * intercept + slope

# Predict weight for a given height
test_height = int(input("Enter a sample height: "))
pred_weight = pred_y(test_height)

# Plot the data and regression line
plt.title("Linear Regression")
plt.xlabel("Height in CM")
plt.ylabel("Weight in KG")
plt.scatter(x, y, color='green', label='Data')
plt.plot(x, pred_y(x), color='red', label='Regression Line')
plt.scatter(test_height, pred_weight, color='#665191', label='Predicted Value')
plt.legend()
plt.show()

# Print the predicted weight
print(f'Predicted weight for a height of {test_height} cm is {pred_weight:.2f} kg')
```
